## Import packages

````python
import os
import numpy as np

import matplotlib # To plot images
import matplotlib.pyplot as plt # Plotting
from matplotlib.colors import LogNorm # Look up table

font = {'family' : 'serif',
            'size'   : 20
        }
matplotlib.rc('font', **font)

# Uncomment the line below to use LaTeX fonts
# matplotlib.rc('text', usetex=True)

from gvxrPython3 import gvxr # Simulate X-ray images
from gvxrPython3.utils import loadSpectrumSpekpy, getSpectrumSpekpy
````

````python
log_norm_min_val = 0.4
energy_bin_size_in_keV = 3
````

````python
# %load snippets/polychromatism-01.py
log_norm_min_val = 1e-6
````

Where to save the plots

````python
output_path = "../notebooks/output_data/polychromatism"
if not os.path.exists(output_path):
    os.makedirs(output_path)
````

Create an OpenGL context

````python
gvxr.createOpenGLContext()
````

Set up the detector

````python
gvxr.setDetectorPosition(10.0, 0.0, 0.0, "cm")
gvxr.setDetectorUpVector(0, 0, -1)
gvxr.setDetectorNumberOfPixels(640, 320)
gvxr.setDetectorPixelSize(0.5, 0.5, "mm")
````

Load the sample data

````python
# Locate the sample STL file from the package directory
path = os.path.dirname(gvxr.__file__)
fname = path + "/welsh-dragon-small.stl"

# Make sure it exists
if not os.path.exists(fname):
    raise IOError(fname)

# Load the file
gvxr.loadMeshFile("Dragon", fname, "mm")

# Move its bounding box centre to (0, 0, 0)
gvxr.moveToCentre("Dragon")

# Material property: Liquid water
gvxr.setCompound("Dragon", "H2O")
gvxr.setDensity("Dragon", 1.0, "g.cm-3")

# Material property: Brass
# gvxr.setMixture("Dragon", ["Cu", "Zn"], [0.66, 0.44])
# gvxr.setDensity("Dragon", 8.73, "g.cm-3")
````

````python
# %load 'snippets/polychromatism-02.py'
# Material property: Brass
gvxr.setMixture("Dragon", ["Cu", "Zn"], [0.66, 0.44])
gvxr.setDensity("Dragon", 8.73, "g.cm-3")
````

Create a source

````python
gvxr.setSourcePosition(-40.0,  0.0, 0.0, "cm")
gvxr.usePointSource()
gvxr.setEnergyBinSize(energy_bin_size_in_keV, "keV")
````

Set the tube voltage

```python
gvxr.setVoltage(120, "kV")
```

````python
kV = 120
gvxr.setVoltage(kV, "kV")
````

## No filtration at all

Plot the corresponding spectrum

````python
energy_bins_no_filter = gvxr.getEnergyBins("keV")
photon_count_no_filter = np.array(gvxr.getPhotonCountsPerPixelAtSDD(), dtype=np.single)
photon_count_no_filter = np.array(gvxr.getPhotonCountsPerCm2At1m(), dtype=np.single)
photon_count_no_filter /= photon_count_no_filter.sum()

plt.figure(figsize=(20, 10))
plt.bar(energy_bins_no_filter, photon_count_no_filter, width=energy_bin_size_in_keV)
plt.xlabel('Energy in keV')
plt.ylabel('Probability distribution of photons per keV')
plt.xlim((0, kV))
plt.tight_layout()
plt.savefig(output_path + "/spectrum-" + str(kV) + "kV-no_filtration.pdf")
plt.savefig(output_path + "/spectrum-" + str(kV) + "kV-no_filtration.png")
plt.show()
````

Compute an X-ray image

````python
x_ray_image_no_filtration = np.array(gvxr.computeXRayImage(), dtype=np.single) / gvxr.getTotalEnergyWithDetectorResponse()
````

Plot the image. This is a naive approach to setting the X-ray tube spectrum if one wants to replicate real-life
experimental conditions.

````python
plt.figure(figsize=(17, 7.5))

plt.imshow(x_ray_image_no_filtration, norm=LogNorm(vmin=log_norm_min_val, vmax=1.0), cmap="gray")
plt.colorbar(orientation='horizontal')
plt.title("Without filtration")
````

## Add the inherent filters that real-life devices will have

It is important to add the inherent filtration, which corresponds to

1. a bit of material by the X-ray tube (e.g. aluminium), and
2. a plate to protect the detector (e.g. carbon).

````python
gvxr.addInherentFilter("Al", 0.5, "mm")
gvxr.addInherentFilter("C", 5, "mm")
````

Plot the spectrum

````python
energy_bins_with_inherent_filter = gvxr.getEnergyBins("keV")
photon_count_with_inherent_filter = np.array(gvxr.getPhotonCountsPerPixelAtSDD(), dtype=np.single)
photon_count_with_inherent_filter = np.array(gvxr.getPhotonCountsPerCm2At1m(), dtype=np.single)
photon_count_with_inherent_filter /= photon_count_with_inherent_filter.sum()

plt.figure(figsize=(20, 10))
plt.bar(energy_bins_with_inherent_filter, photon_count_with_inherent_filter, width=energy_bin_size_in_keV)
plt.xlabel('Energy in keV')
plt.ylabel('Probability distribution of photons per keV')
plt.xlim((0, kV))
plt.tight_layout()
plt.savefig(output_path + "/spectrum-" + str(kV) + "kV-with_inherent_filter.pdf")
plt.savefig(output_path + "/spectrum-" + str(kV) + "kV-with_inherent_filter.png")
plt.show()
````

Compute an X-ray image

````python
x_ray_image_with_inherent_filtration = (np.array(gvxr.computeXRayImage(), dtype=np.single) / gvxr
                                .getTotalEnergyWithDetectorResponse());
````

Plot the image. You can already see an improvement as low-energy photons were absorbed by the inherent filtration.
This is more realistic.

````python
plt.figure(figsize=(17, 7.5))

plt.imshow(x_ray_image_with_inherent_filtration, norm=LogNorm(vmin=log_norm_min_val, vmax=1.0), cmap="gray")
plt.colorbar(orientation='horizontal')
plt.title("With inherent filtration")
````

## Add a copper filter in addition to the inherent filters

Add a copper filter

````python
gvxr.addFilter("Cu", 2.0, "mm")
````

````python
# %load 'snippets/polychromatism-03.py'
gvxr.addFilter("Cu", 3.0, "mm")
````

Plot the spectrum

````python
energy_bins_with_filter = gvxr.getEnergyBins("keV")
photon_count_with_filter = np.array(gvxr.getPhotonCountsPerPixelAtSDD(), dtype=np.single)
photon_count_with_filter = np.array(gvxr.getPhotonCountsPerCm2At1m(), dtype=np.single)
photon_count_with_filter /= photon_count_with_filter.sum()

plt.figure(figsize=(20, 10))
plt.bar(energy_bins_with_filter, photon_count_with_filter, width=energy_bin_size_in_keV)
plt.xlabel('Energy in keV')
plt.ylabel('Probability distribution of photons per keV')
plt.xlim((0, kV))
plt.tight_layout()
plt.savefig(output_path + "/spectrum-" + str(kV) + "kV-with_all_filters.pdf")
plt.savefig(output_path + "/spectrum-" + str(kV) + "kV-with_all_filters.png")
plt.show()
````

Compute an X-ray image

````python
x_ray_image_with_filtration = np.array(gvxr.computeXRayImage(), dtype=np.single) / gvxr.getTotalEnergyWithDetectorResponse();
````

Plot the image. You can already see a further improvement. Adding filters is used in tomography to minimise beam
hardening artefacts.

````python
plt.figure(figsize=(17, 7.5))

plt.imshow(x_ray_image_with_filtration, norm=LogNorm(vmin=log_norm_min_val, vmax=1.0), cmap="gray")
plt.colorbar(orientation='horizontal')
plt.title("With all filtration")
````

Compare the spectra

````python
plt.figure(figsize=(17, 7.5))

plt.suptitle("Spectrum comparison")

plt.subplot(131)
plt.bar(energy_bins_no_filter, photon_count_no_filter, width=energy_bin_size_in_keV)
plt.title("Without filtration")
plt.xlim((0, kV))
plt.xlabel('Energy in keV')
plt.ylabel('Probability distribution\nof photons per keV')

plt.subplot(132)
plt.bar(energy_bins_with_inherent_filter, photon_count_with_inherent_filter, width=energy_bin_size_in_keV)
plt.title("With inherent filtration")
plt.xlim((0, kV))
plt.xlabel('Energy in keV')

plt.subplot(133)
plt.bar(energy_bins_with_filter, photon_count_with_filter, width=energy_bin_size_in_keV)
plt.title("With all filters")
plt.xlim((0, kV))
plt.xlabel('Energy in keV')

plt.tight_layout()
````

Display the X-ray images and compare them. It is possible to see that the contrast has improved.

````python
matplotlib.rcParams['xtick.labelsize'] = 'xx-small'
matplotlib.rcParams['ytick.labelsize'] = 'xx-small'

plt.figure(figsize=(17, 7.5))

plt.suptitle("Image comparison", y=0.75)

plt.subplot(131)
plt.imshow(x_ray_image_no_filtration, norm=LogNorm(vmin=log_norm_min_val, vmax=1.0), cmap="gray")
cbar1 = plt.colorbar(orientation='horizontal')
plt.title("Without filtration")

plt.subplot(132)
plt.imshow(x_ray_image_with_inherent_filtration, norm=LogNorm(vmin=log_norm_min_val, vmax=1.0), cmap="gray")
cbar2 = plt.colorbar(orientation='horizontal')
plt.title("With inherent filtration")

plt.subplot(133)
plt.imshow(x_ray_image_with_filtration,  norm=LogNorm(vmin=log_norm_min_val, vmax=1.0), cmap="gray")
cbar3 = plt.colorbar(orientation='horizontal')
plt.title("With all filters")

plt.tight_layout()
````

Plot the corresponding diagonal profiles

````python
matplotlib.rcParams['xtick.labelsize'] = 'medium'
matplotlib.rcParams['ytick.labelsize'] = 'medium'

plt.figure(figsize=(17, 7.5))

plt.suptitle("Diagonal profiles")

plt.plot(np.diag(x_ray_image_no_filtration), label="Without filtration")
plt.plot(np.diag(x_ray_image_with_inherent_filtration), label="With inherent filtration")
plt.plot(np.diag(x_ray_image_with_filtration), label="With all filters")
plt.legend()
plt.xlim((125, 275))
plt.tight_layout()
````

The intensity histogram of an image serves as a powerful visual and statistical indicator of the image's contrast by
mapping the frequency of pixels at each grayscale brightness level. As a rule of thumb, the flatter the histogram, th
better the contrast.

````python
plt.figure(figsize=(17, 7.5))

n_bins = 50

plt.suptitle("Image histograms")

plt.subplot(131)
plt.title("Without filtration")
plt.xlabel('Intensity')
plt.ylabel('Frequency')
plt.hist(x_ray_image_no_filtration.ravel(), bins=n_bins, range=(0.0, 0.99))
# plt.yscale("log", nonpositive='clip')

plt.subplot(132)
plt.title("With inherent filtration")
plt.xlabel('Intensity')
plt.hist(x_ray_image_with_inherent_filtration.ravel(), bins=n_bins, range=(0.0, 0.99))
# plt.yscale("log", nonpositive='clip')

plt.subplot(133)
plt.title("With all filters")
plt.xlabel('Intensity')
plt.hist(x_ray_image_with_filtration.ravel(), bins=n_bins, range=(0.0, 0.99))
# plt.yscale("log", nonpositive='clip')
plt.tight_layout()
````

The cumulative histogram is maybe a better indicator of poor/good contrast. It is a monotonic function as it can only
increase as each histogram bin. Poor contrast will correspond to cumulative histograms that are fairly horizontal.
When the cumulative histogram is a ramp, good contrast will be observed in the image.

````python
plt.figure(figsize=(17, 7.5))

n_bins = 100

plt.suptitle("Image histograms")

plt.subplot(131)
plt.title("Without filtration")
plt.xlabel('Intensity')
plt.ylabel('Frequency')
plt.hist(x_ray_image_no_filtration.ravel(),
         bins=n_bins, range=(0.0, 0.99), density=True, cumulative=True)

plt.subplot(132)
plt.title("With inherent filtration")
plt.xlabel('Intensity')
plt.hist(x_ray_image_with_inherent_filtration.ravel(),
         bins=n_bins, range=(0.0, 0.99), density=True, cumulative=True)

plt.subplot(133)
plt.title("With all filters")
plt.xlabel('Intensity')
plt.hist(x_ray_image_with_filtration.ravel(),
         bins=n_bins, range=(0.0, 0.99), density=True, cumulative=True)
plt.tight_layout()
````

The improvement of contrast was limited. This is because the sample material was not very dense.
